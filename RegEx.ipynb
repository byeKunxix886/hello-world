**[^(字元集)]** 限定除了(字元集)以外的，如果打^K表示除了K以外的都要，包含數字0-9、white-space、k，若不要這些也要加以限制。

1.開頭是K的身分證字號不要。[^K][0-9][0-9][0-9][0-9][0-9][0-9][0-9][0-9][0-9]


**.** 任何字元(except newline character)

```python
import re

def all_char():
  str1 = "K123215205"

  x_rg = re.findall("[^K]",str1)

  return x_rg

all_char()
```
